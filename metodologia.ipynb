```python
import numpy as np
import pandas as pd
import random as rd
```

```python
df_capacity01 = pd.read_excel(
    '../Data/Estimación de aforo para la realización de un evento (1)(1-47).xlsx'
)
```

```python
df_capacity02 = pd.read_excel(
    '../Data/Estimación de aforo para la realización de un evento (2)(1-33).xlsx'
)
```

```python
df_capacity03 = pd.read_excel(
    '../Data/Estimación de aforo para la realización de un evento (3)(1-27).xlsx'
)
```

```python
def fill_nulls_recomendacion(row):
    if pd.isnull(row['recomendacion']):
        row['recomendacion'] = rd.choice(['Si', 'No'])
        return row['recomendacion']
    else:
        return row['recomendacion']
```

Limpieza del grupo experimental 2

```python
df_capacity01
```

```python
df_capacity01 = df_capacity01.iloc[:, 5:]
```

```python
df_capacity01.columns = ['valor_estimado_1', 'confianza_1', 'valor_estimado_2', 'confianza_2', 'valor_estimado_3', 'confianza_3', 'valor_entre', 'confianza_entre', 'recomendacion']
```

```python
df_capacity01 = df_capacity01[(df_capacity01['valor_estimado_1'] > 10) & (df_capacity01['valor_estimado_2'] > 10) & (df_capacity01['valor_estimado_3'] > 10)]
```

```python
df_capacity01 = df_capacity01.sample(n=30).reset_index(drop=True)
```

```python
df_capacity01['valor_entre'] = df_capacity01['valor_entre'].str.replace('Entre ', '').str.replace(' y ', ' ').str.replace(' personas', '').str.replace('entre ', '').str.replace('-', '').str.replace('unas ', '').str.replace(' a ', ' ').str.replace('    ', ' ')
```

```python
def append_upper_bound(row):
    if len(row['valor_entre'].split(' ')) == 1:
        return row['valor_entre'] + ' ' + str((int(row['valor_entre']) + 15))
    else:
        return row['valor_entre']
```

```python
df_capacity01['valor_entre'] = df_capacity01.apply(append_upper_bound, axis=1, result_type='expand')
```

```python
df_capacity01[['valor_desde', 'valor_hasta']] = df_capacity01['valor_entre'].str.split(' ', expand=True)
```

```python
df_capacity01.drop(columns=['valor_entre'], inplace=True)
```

```python
df_capacity01['valor_desde'] = df_capacity01['valor_desde'].astype(int)
df_capacity01['valor_hasta'] = df_capacity01['valor_hasta'].astype(int)
```

```python
def fill_nulls_confianza_entre(row):
    if pd.isnull(row['confianza_entre']):
        row['confianza_entre'] = rd.randint(10, 100)
        return row['confianza_entre']
    else:
        return row['confianza_entre']
```

```python
df_capacity01['confianza_entre'] = df_capacity01.apply(fill_nulls_confianza_entre, axis=1, result_type='expand')
```

```python
df_capacity01['recomendacion'] = df_capacity01.apply(fill_nulls_recomendacion, axis=1, result_type='expand')
```
